```python
import pandas as pd
```

```python
df=pd.read_csv('merge-csv.com__6490b905d26ef.csv', encoding= 'unicode_escape')
df
```

```python
word_list={}
word_list['Travel']=travel
word_list['Dining']=dining
word_list['Visits']=visits
word_list['Business']=business
```

```python
dataf={}
for f in word_list:
    dataf[f]=[]
    for j in df['sentences']:
        for word in word_list[f]:
            if word in j.lower():
                dataf[f].append(j.lower())
                break
                
```

```python

pir=list(dataf.keys())
#lenpir=[len(dataf[key]) for key in pir]
#
sizes=lenpir
```

```python
labels=pir
sizes=lenpir
patterns = ['/', '\\', 'x', 'o', '+']  # Shading patterns for each slice
colors = ['yellow', '#F0F8FF', '#FFEFDB', 'orange']  # Colors for each pattern

# Create the pie chart
fig, ax = plt.subplots()
wedges, _, autotexts = ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90,
                              wedgeprops={'edgecolor': 'black'},textprops={'fontsize': 15})  # Set edge color

# Set color and pattern for each wedge
for wedge, pattern, color, autotext in zip(wedges, patterns, colors, autotexts):
    wedge.set_color(color)
    autotext.set_color('black')  # Set color for text labels
     # Set pattern
    wedge.set_edgecolor('black')  # Set edge color
    
    #wedge.set_hatch(pattern)

   

# Set aspect ratio to be equal so that pie is drawn as a circle
ax.axis('equal')

plt.show()
```
